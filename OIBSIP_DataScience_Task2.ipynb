**Unemployment Analysis with Python**

*Import Libraries*

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

*Load  Dataset*

```python
df = pd.read_csv("OIBSIP_DataScience_Task2.csv")
```

*Dataset Information*

```python
print("First 5 Rows")
print(df.head())

print("\nDataset Shape")
print(df.shape)

```

*Check Missing Values*

```python
print("\nMissing Values")
print(df.isnull().sum())
```

*Clean Column Names*

```python
df.columns = df.columns.str.strip()

print("\nColumns:")
print(df.columns)
```

*Convert Date*

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
```

*Average Unemployment Rate*

```python
print("\nAverage Unemployment Rate:")
print(df['Estimated Unemployment Rate (%)'].mean())
```

*State-Wise Analysis*

```python
state_unemployment = df.groupby('Region')['Estimated Unemployment Rate (%)'].mean()

print("\nState Wise Average Unemployment")
print(state_unemployment.sort_values(ascending=False))
```

*Line Graph*

```python
plt.figure(figsize=(12,6))

sns.lineplot(
    data=df,
    x='Date',
    y='Estimated Unemployment Rate (%)'
)

plt.title("Unemployment Rate Over Time")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate")

plt.xticks(rotation=45)

plt.show()



```

*Bar Graph*

```python
top_states = state_unemployment.sort_values(ascending=False).head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    x=top_states.values,
    y=top_states.index
)

plt.title("Top 10 States by Unemployment Rate")
plt.xlabel("Average Unemployment Rate")
plt.ylabel("State")

plt.show()
```

**Conclusion**

The unemployment rate increased significantly during the Covid-19 period. Some states showed consistently higher unemployment rates compared to others. Data visualization helped identify unemployment trends and regional differences effectively.
